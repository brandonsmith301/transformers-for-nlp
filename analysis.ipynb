```python
import pandas as pd
import json
import pprint as pp
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# --- bert base uncased --- #
with open("code/outputs/bert/bert_base_uncased/logs/best_model_log.json", "r") as f:
    data = json.load(f)

dict_list = []

for i in range(len(data["training_history"])):
    dict_metrics = {}
    dict_metrics["model"] = data["model_type"]
    dict_metrics["step"] = data["training_history"][i]["total_step"]
    dict_metrics["time"] = data["training_history"][i]["time_elapsed"]
    dict_metrics["accuracy"] = data["training_history"][i]["accuracy"]
    dict_list.append(dict_metrics)

bert = pd.DataFrame(dict_list)
```

```python
# --- lstm --- #
with open("code/outputs/lstm/lstm/logs/best_model_log.json", "r") as f:
    data = json.load(f)

dict_list = []

for i in range(len(data["training_history"])):
    dict_metrics = {}
    dict_metrics["model"] = data["model_type"]
    dict_metrics["step"] = data["training_history"][i]["total_step"]
    dict_metrics["time"] = data["training_history"][i]["time_elapsed"]
    dict_metrics["accuracy"] = data["training_history"][i]["accuracy"]
    dict_list.append(dict_metrics)

lstm = pd.DataFrame(dict_list)
```

```python
# --- custom models --- #
all_dfs = []
for model in os.listdir("code/outputs/custom"):
    with open(f"code/outputs/custom/{model}/logs/best_model_log.json", "r") as f:
        data = json.load(f)

    dict_list = []

    for i in range(len(data["training_history"])):
        dict_metrics = {}
        dict_metrics["model"] = model
        dict_metrics["step"] = data["training_history"][i]["total_step"]
        dict_metrics["time"] = data["training_history"][i]["time_elapsed"]
        dict_metrics["accuracy"] = data["training_history"][i]["accuracy"]
        dict_list.append(dict_metrics)

    df = pd.DataFrame(dict_list)
    all_dfs.append(df)

# add bert
all_dfs.append(bert)

# add lstm
all_dfs.append(lstm)
```

```python
all_dfs = pd.concat(all_dfs)
all_dfs = all_dfs.sort_values("accuracy", ascending=False)
all_dfs = all_dfs.drop_duplicates(subset="model", keep="first")
```
